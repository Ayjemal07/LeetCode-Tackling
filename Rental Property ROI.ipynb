```python
class RentalProperty():
    
    def __init__(self):
        pass
    
    def ask_input(self):
        while True:
            self.income=input("What is your total monthly income? This is your rental/storage/laundry income all combined ")
            if self.income.isdigit():
                break
            elif self.income.lower()=="quit":
                print("Program Ended. Thanks for using ROI calculation with us!")
                return False
            else:
                print("Invalid Input, please enter integers only without dollar sign and try again.")
                
        while True:
            self.expenses=input("What is your total monthly expense? This includes taxes/electric etc. ")
            if self.expenses.isdigit():
                break
            elif self.expenses.lower()=="quit":
                print("Program Ended. Thanks for using ROI calculation with us!")
                return False
            else:
                print("Invalid Input, please enter integers only without dollar sign and try again.")
                
        while True:
            self.investment=input("What is the total amount you invested on this property including down payment/closing cost etc ? ")
            if self.investment.isdigit():
                break
            elif self.investment.lower()=="quit":
                print("Program Ended. Thanks for using ROI calculation with us!")
                return False
            else:
                print("Invalid Input, please enter integers only without dollar sign and try again.")        
        
    def calculation(self):
        self.cash_flow=(int(self.income)-int(self.expenses))*12
        self.roi=self.cash_flow/int(self.investment)
        self.percent=(self.roi*100)
        print(f"Your ROI is {self.percent}")

    def calculateRoi(self):
        print("Welcome to ROI Calculator! You would need total monthly income, expenses and investment handy to start this calculation. Let's start!")
        while True:
            if self.ask_input()==False:
                break
            self.calculation()

            prompt=input('Would you like to calculate ROI for another property? Type "yes" or "no" below ')
            if prompt.lower()=="yes":
                continue
            elif prompt.lower()=="no" or prompt.lower()=="quit":
                print("Thanks for using our ROI calculation! Have a good day!")
                break
            else:
                print("Please type valid prompt")            

RentalProperty().calculateRoi()
```
