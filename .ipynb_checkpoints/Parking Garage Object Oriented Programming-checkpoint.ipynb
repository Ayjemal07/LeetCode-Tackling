```python
"""
Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 


Your parking garage class should have the following methods:

- takeTicket
- This should decrease the amount of tickets available by 1
- This should decrease the amount of parkingSpaces available by 1

- payForParking
- Display an input that waits for an amount from the user and store it in a variable
- If the payment variable is not empty then (meaning the ticket has been paid) -> 
    display a message to the user that their ticket has been paid and they have 15mins to leave
- This should update the "currentTicket" dictionary key "paid" to True

-leaveGarage
- If the ticket has been paid, display a message of "Thank You, have a nice day"
- If the ticket has not been paid, display an input prompt for payment
- Once paid, display message "Thank you, have a nice day!"
- Update parkingSpaces list to increase by 1 (meaning add to the parkingSpaces list)
- Update tickets list to increase by 1 (meaning add to the tickets list)

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

By the end of this project each student should be able to:
- Explain and/or demonstrate creating classes
- Explain and/or demonstrate creating class methods
- Explain and/or demonstrate class instantiation
"""

class ParkingGarage():
    
    
    def __init__(self):
        self.capacity = 7
        self.tickets=[0]*self.capacity
        self.parkingSpaces=[1]*self.capacity
        self.currentTicket={"paid":False,"taken":False}
    
    def takeTicket(self):
        if len(self.tickets)>=1:
            self.tickets.pop()
            self.parkingSpaces.pop()
            self.currentTicket["taken"]=True
            print("Your ticket has been issued!")
        else:
            print("We don't have any parking spaces available at the moment, check back later ")
    
    def payforParking(self):
        payment=input("Enter '$10' to pay for this ticket ")
        
        if payment=="$10" or payment=="10":
            print("Your ticket has been paid, you have 15 min to leave ")
            self.currentTicket["paid"]=True
            
        else:
            print("Invalid input try again")
            self.payforParking()
    
    def leaveGarage(self):
        print("Thank You, have a nice day")
        self.tickets.append(0)
        self.parkingSpaces.append(1)

    def run(self):
        while True:
            prompt=input("Type 'take' for a ticket, 'pay' to pay for ticket, 'leave' to leave the garage ")

            if prompt=="take":
                self.takeTicket()

            elif prompt=="pay":
                if self.currentTicket["taken"]==True:
                    self.payforParking()
                else:
                    print("You need to take a ticket before paying ")
                    
            elif prompt=="leave":
                if self.currentTicket["taken"]==True and self.currentTicket["paid"]==True:
                    self.leaveGarage()
                    break
                else:
                    print("Thanks for visiting! You need to take a ticket, pay for it in order to exit the garage after you enter ")
            
            else:
                print("Invalid input, try again")
                                      
myGarage=ParkingGarage()
myGarage.run()
```
