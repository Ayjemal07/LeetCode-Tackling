# Working with APIs

API stands for Application programming interface<br>
APIs are a way for applications to talk to eachother<br>
In python we use the requests package to handle the connection<br>
<small><strong>Note: It is requests with an s not request(this is a different package)</strong></small><br>
Most API's communicate with data in the form of JSON<br>
JSON stands for JavaScript Object Notation

```python
#import the packages we need:
import requests, json
```

We will be connecting to the Ergast F1 Racer API today:
http://ergast.com/mrd/

to view JSON data nicely in the Chrome browser install the extension JSONview

```python
# help(requests)
```

```python
url = 'http://ergast.com/api/f1/2008/5/driverStandings.json'
```

```python
#response object

response=requests.get(url)
```

```python
#check and make sure we got a successful response from the API
print(response)
```

```python
#view the response as a JSON
response.json()
```

```python
type(response.json())
```

```python
# We only want the Driver Standings
data=response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][0]
data
```

```python
pokemon_data=requests.get("https://pokeapi.co/api/v2/pokemon/ditto")


for i in range(1,21):
    pokemon_data=requests.get(f"https://pokeapi.co/api/v2/pokemon/{i}")
    data=pokemon_data.json()
    name=data["name"]
    ability=data["abilities"][0]["ability"]["name"]
    base_experience=data["base_experience"]
    print(F"My name is: {name}. One of my abilites is {ability}. Base experience: {base_experience}")

    url_sprite=data["sprites"]["front_shiny"]
    print(F"Find my front shiny sprite below: \n{url_sprite}")

    hp=data["stats"][0]["base_stat"]

    attack=data["stats"][1]["base_stat"]

    defense=data["stats"][2]["base_stat"]

    print(F"My hp base stat is {hp}\nMy attack base is {attack}\nFinally, my defense base is {defense}\n")
```
